# Loading the book

```python
filename = "miracle_in_the_andes.txt"
with open(filename, "r", encoding="utf-8") as file:
    book = file.read()
```

# Most used words (non-articles)

```python
import re
import nltk
from nltk.corpus import stopwords
```

## Function to count instances of each word (no stopwords)

```python
def count_word_instances(text):
    english_stopwords = stopwords.words("english")
    pattern = re.compile("[a-zA-z]+")
    findings = re.findall(pattern, text)
    d = {}
    for word in findings:
        if word.lower() in english_stopwords:
            pass
        elif word in d.keys():
            d[word] = d[word] + 1
        else:
            d[word] = 1
    word_list = [(value, key) for (key, value) in d.items()]
    word_list = sorted(word_list, reverse=True)
    return word_list
```

```python
book_words = count_word_instances(book)
```

# Sentiment analysis: find the most positive and most negative chapters

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

## Analyze by chapter

```python
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)[1:]
```

```python
for num, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    print(num + 1, scores)
```
